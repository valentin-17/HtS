# Imports

```python
from datetime import date, timedelta

import polars as pl
import holidays
import matplotlib.pyplot as plt
import random
```

# External Data Prep

```python
semester_terms = [
    {
        "semester": "winter_2024_2025",
        "semester_season": "winter",
        "semester_start": date(2024, 10, 1),
        "semester_end": date(2025, 3, 31),
        "lecture_start": date(2024, 10, 21),
        "lecture_end": date(2025, 2, 15),
    },
    {
        "semester": "summer_2025",
        "semester_season": "summer",
        "semester_start": date(2025, 4, 1),
        "semester_end": date(2025, 9, 30),
        "lecture_start": date(2025, 4, 14),
        "lecture_end": date(2025, 7, 19),
    },
    {
        "semester": "winter_2025_2026",
        "semester_season": "winter",
        "semester_start": date(2025, 10, 1),
        "semester_end": date(2026, 3, 31),
        "lecture_start": date(2025, 10, 13),
        "lecture_end": date(2026, 2, 14),
    },
]

public_holidays = holidays.country_holidays("DE", subdiv="RP", years=[2025])
public_holiday_rows = [
    {"day": holiday_day, "public_holiday_name": holiday_name}
    for holiday_day, holiday_name in sorted(public_holidays.items())
]
public_holiday_df = pl.DataFrame(
    public_holiday_rows,
    schema={"day": pl.Date, "public_holiday_name": pl.Utf8},
)

university_holidays = {
    "winter_2024_2025_christmas_break": (
        date(2024, 12, 23),
        date(2025, 1, 5),
    ),
    "winter_2025_2026_christmas_break": (
        date(2025, 12, 22),
        date(2026, 1, 7),
    ),
    # "summer_2025_university_holiday": (
    #     date(2025, 8, 1),
    #     date(2025, 8, 15),
    # ),
}

university_holiday_rows = [
    {
        "day": holiday_start + timedelta(days=offset),
        "university_holiday_name": description,
    }
    for description, (holiday_start, holiday_end) in university_holidays.items()
    for offset in range((holiday_end - holiday_start).days + 1)
]
university_holiday_df = pl.DataFrame(
    university_holiday_rows,
    schema={"day": pl.Date, "university_holiday_name": pl.Utf8},
)

campus_events = {
    'Tagung "Aufklaerung jenseits der Oeffentlichkeit"': (
        date(2025, 9, 4),
        date(2025, 9, 5),
    ),
}

campus_event_rows = [
    {
        "day": event_start + timedelta(days=offset),
        "campus_event_description": description,
    }
    for description, (event_start, event_end) in campus_events.items()
    for offset in range((event_end - event_start).days + 1)
]
campus_events_df = pl.DataFrame(
    campus_event_rows,
    schema={"day": pl.Date, "campus_event_description": pl.Utf8},
)


def period_label(periods: list[dict], label_col: str, start_col: str, end_col: str) -> pl.Expr:
    expr = pl.lit(None, dtype=pl.Utf8)
    for period in periods:
        expr = pl.when(pl.col("day").is_between(period[start_col], period[end_col], closed="both"))\
            .then(pl.lit(period[label_col]))\
            .otherwise(expr)
    return expr


def in_period(periods: list[dict], start_col: str, end_col: str) -> pl.Expr:
    expr = pl.lit(False)
    for period in periods:
        expr = expr | pl.col("day").is_between(period[start_col], period[end_col], closed="both")
    return expr


def in_relative_lecture_start_window(start_offset_days: int, end_offset_days: int) -> pl.Expr:
    expr = pl.lit(False)
    for period in semester_terms:
        window_start = period["lecture_start"] + timedelta(days=start_offset_days)
        window_end = period["lecture_start"] + timedelta(days=end_offset_days)
        expr = expr | pl.col("day").is_between(window_start, window_end, closed="both")
    return expr


calendar_2025 = (
    pl.DataFrame(
        {
            "day": pl.date_range(
                date(2025, 1, 1),
                date(2025, 12, 31),
                interval="1d",
                eager=True,
            )
        }
    )
    .with_columns(
        weekday_number=pl.col("day").dt.weekday(),  # Monday=1, Sunday=7
        weekday=pl.col("day").dt.strftime("%A"),
        iso_week=pl.col("day").dt.week(),
        month=pl.col("day").dt.month(),
        quarter=pl.col("day").dt.quarter(),
        day_of_year=pl.col("day").dt.ordinal_day(),
    )
    .with_columns(
        is_weekend=pl.col("weekday_number").is_in([6, 7]),
        season=pl.when(pl.col("month").is_in([3, 4, 5]))
        .then(pl.lit("spring"))
        .when(pl.col("month").is_in([6, 7, 8]))
        .then(pl.lit("summer"))
        .when(pl.col("month").is_in([9, 10, 11]))
        .then(pl.lit("autumn"))
        .otherwise(pl.lit("winter")),
        semester=period_label(semester_terms, "semester", "semester_start", "semester_end"),
        semester_season=period_label(semester_terms, "semester_season", "semester_start", "semester_end"),
        is_semester_day=in_period(semester_terms, "semester_start", "semester_end"),
        is_lecture_day=in_period(semester_terms, "lecture_start", "lecture_end"),
        is_orientation_week=in_relative_lecture_start_window(-7, -1),
        is_first_lecture_week=in_relative_lecture_start_window(0, 6),
    )
    .with_columns(
        is_semester_start_period=pl.col("is_orientation_week") | pl.col("is_first_lecture_week"),
    )
    .join(public_holiday_df, on="day", how="left")
    .join(university_holiday_df, on="day", how="left")
    .join(campus_events_df, on="day", how="left")
    .with_columns(
        is_public_holiday=pl.col("public_holiday_name").is_not_null(),
        is_university_holiday=pl.col("university_holiday_name").is_not_null(),
        is_campus_event=pl.col("campus_event_description").is_not_null(),
    )
    .with_columns(
        is_university_closure_proxy=pl.col("is_public_holiday") | pl.col("is_university_holiday"),
    )
    .with_columns(
        academic_bucket=pl.when(pl.col("is_university_closure_proxy"))
        .then(pl.lit("university_closure"))
        .when(pl.col("is_semester_start_period"))
        .then(pl.lit("semester_start_period"))
        .when(pl.col("is_lecture_day"))
        .then(pl.lit("lecture"))
        .when(pl.col("is_semester_day"))
        .then(pl.lit("semester_lecture_free"))
        .otherwise(pl.lit("error")),
        is_cafeteria_relevant_day=~pl.col("is_weekend"),
        is_regular_service_day=(~pl.col("is_weekend")) & (~pl.col("is_university_closure_proxy")),
    )
)

# Keep closure days visible in the main weekday calendar so their flags can be inspected.
cafeteria_calendar_2025 = calendar_2025.filter(pl.col("is_cafeteria_relevant_day"))

# Use this for analyses that should exclude public/university closure days.
regular_service_calendar_2025 = calendar_2025.filter(pl.col("is_regular_service_day"))

calendar_2025
```

# Mock Sales Data

```python
MEAL_CATEGORIES = [
    "Sides",
    "Sides/Salads",
    "Sides/Starch",
    "Bowls - Oliva",
    "Soup",
    "Joker",
    "Menu 1",
    "Menu 2",
    "Seconds",
    "Pasta - Oliva",
    "Pizza - Oliva",
    "Saladplate",
    "Sweet Joker",
    "Toppings - Oliva",
]

MOCK_RANDOM_SEED = 42
MOCK_DAILY_TOTAL_BASE = 1500
MOCK_DAILY_TOTAL_RANGE = (400, 2000)
MOCK_DAILY_TOTAL_VARIABILITY = 0.10
MOCK_CATEGORY_CONCENTRATION = 120.0
MOCK_MINOR_CATEGORY_ZERO_PROBABILITY = 0.10

MOCK_CATEGORY_FRACTIONS = {
    "Menu 1": 0.22,
    "Menu 2": 0.18,
    "Pizza - Oliva": 0.14,
    "Pasta - Oliva": 0.13,
    "Sides/Salads": 0.11,
    "Sweet Joker": 0.08,
    "Sides/Starch": 0.035,
    "Soup": 0.025,
    "Joker": 0.025,
    "Saladplate": 0.02,
    "Bowls - Oliva": 0.015,
    "Sides": 0.01,
    "Seconds": 0.005,
    "Toppings - Oliva": 0.005,
}

MOCK_WEEKDAY_MULTIPLIERS = {
    "Monday": 0.96,
    "Tuesday": 1.03,
    "Wednesday": 1.06,
    "Thursday": 1.02,
    "Friday": 0.90,
}

MOCK_ACADEMIC_BUCKET_MULTIPLIERS = {
    "semester_start_period": 1.10,
    "lecture": 1.00,
    "semester_lecture_free": 0.6,
    "university_closure": 0.00,
}

MOCK_DISH_NAMES = {
    "Menu 1": ["Menu 1 Classic", "Menu 1 Curry", "Menu 1 Ragout", "Menu 1 Roast"],
    "Menu 2": ["Menu 2 Veggie", "Menu 2 Lentils", "Menu 2 Gratin", "Menu 2 Stir Fry"],
    "Pizza - Oliva": ["Pizza Margherita", "Pizza Verdure", "Pizza Funghi", "Pizza Piccante"],
    "Pasta - Oliva": ["Pasta Pomodoro", "Pasta Pesto", "Pasta Arrabbiata", "Pasta Alfredo"],
    "Sides/Salads": ["Mixed Salad", "Cucumber Salad", "Carrot Salad", "Leaf Salad"],
    "Sweet Joker": ["Rice Pudding", "Pancakes", "Apple Crumble", "Semolina Pudding"],
}


def normalized_category_fractions(fractions: dict[str, float]) -> dict[str, float]:
    missing_categories = set(MEAL_CATEGORIES) - set(fractions)
    if missing_categories:
        raise ValueError(f"Missing category fractions for: {sorted(missing_categories)}")

    total_fraction = sum(fractions.values())
    if total_fraction <= 0:
        raise ValueError("Category fractions must sum to a positive value.")

    return {category: fraction / total_fraction for category, fraction in fractions.items()}


def allocate_integer_total(total: int, weights: dict[str, float]) -> dict[str, int]:
    raw_allocations = {category: total * weight for category, weight in weights.items()}
    allocations = {category: int(value) for category, value in raw_allocations.items()}
    remainder = total - sum(allocations.values())

    remainder_order = sorted(
        raw_allocations,
        key=lambda category: raw_allocations[category] - allocations[category],
        reverse=True,
    )
    for category in remainder_order[:remainder]:
        allocations[category] += 1

    return allocations


def generate_mock_meal_sales(calendar: pl.DataFrame) -> pl.DataFrame:
    rng = random.Random(MOCK_RANDOM_SEED)
    category_fractions = normalized_category_fractions(MOCK_CATEGORY_FRACTIONS)
    core_categories = {
        "Menu 1",
        "Menu 2",
        "Pizza - Oliva",
        "Pasta - Oliva",
        "Sides/Salads",
        "Sweet Joker",
    }

    rows = []
    service_days = calendar.filter(pl.col("is_regular_service_day")).select(
        "day",
        "weekday",
        "academic_bucket",
    )

    for service_day in service_days.iter_rows(named=True):
        weekday_factor = MOCK_WEEKDAY_MULTIPLIERS.get(service_day["weekday"], 1.0)
        academic_factor = MOCK_ACADEMIC_BUCKET_MULTIPLIERS.get(service_day["academic_bucket"], 1.0)
        noise_factor = max(0.70, rng.gauss(1.0, MOCK_DAILY_TOTAL_VARIABILITY))
        daily_total = round(MOCK_DAILY_TOTAL_BASE * weekday_factor * academic_factor * noise_factor)
        daily_total = max(MOCK_DAILY_TOTAL_RANGE[0], min(MOCK_DAILY_TOTAL_RANGE[1], daily_total))

        active_categories = [
            category
            for category in MEAL_CATEGORIES
            if category in core_categories or rng.random() > MOCK_MINOR_CATEGORY_ZERO_PROBABILITY
        ]
        if not active_categories:
            active_categories = ["Menu 1"]

        daily_weights = {}
        for category in active_categories:
            target_fraction = category_fractions[category]
            shape = max(target_fraction * MOCK_CATEGORY_CONCENTRATION, 0.1)
            daily_weights[category] = rng.gammavariate(shape, 1.0)

        weight_sum = sum(daily_weights.values())
        daily_weights = {
            category: weight / weight_sum
            for category, weight in daily_weights.items()
        }
        daily_allocations = allocate_integer_total(daily_total, daily_weights)

        for category, sales in daily_allocations.items():
            if sales == 0:
                continue
            dish_pool = MOCK_DISH_NAMES.get(category, [f"{category} Mock Dish"])
            rows.append(
                {
                    "date": service_day["day"],
                    "meal category": category,
                    "name": rng.choice(dish_pool),
                    "sales": sales,
                }
            )

    return pl.DataFrame(
        rows,
        schema={
            "date": pl.Date,
            "meal category": pl.Utf8,
            "name": pl.Utf8,
            "sales": pl.Int64,
        },
    )


mock_meal_sales = generate_mock_meal_sales(calendar_2025)
mock_daily_total_sales = mock_meal_sales.group_by("date").agg(
    pl.sum("sales").alias("total_sales")
)

mock_daily_total_sales.select(
    service_days=pl.len(),
    min_total_sales=pl.min("total_sales"),
    mean_total_sales=pl.mean("total_sales").round(1),
    max_total_sales=pl.max("total_sales"),
)
```

# Mock Sales Calendar Analysis

```python
mock_sales_with_calendar = mock_meal_sales.join(
    calendar_2025,
    left_on="date",
    right_on="day",
    how="left",
)

daily_total_sales = (
    mock_sales_with_calendar
    .group_by(
        "date",
        "weekday",
        "weekday_number",
        "season",
        "semester",
        "semester_season",
        "is_lecture_day",
        "academic_bucket",
    )
    .agg(pl.sum("sales").alias("total_sales"))
    .sort("date")
    .with_columns(
        pl.col("total_sales")
        .rolling_mean(window_size=10, min_samples=3)
        .alias("rolling_mean_10_service_days")
    )
)

weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
season_order = ["winter", "spring", "summer", "autumn"]
academic_bucket_order = [
    "semester_start_period",
    "lecture",
    "semester_lecture_free",
    "semester_break",
]
semester_order = [term["semester"] for term in semester_terms]


def ordered_bar_data(df: pl.DataFrame, label_col: str, value_col: str, order: list[str]) -> pl.DataFrame:
    order_df = pl.DataFrame(
        {label_col: order, "sort_order": list(range(len(order)))}
    )
    return (
        df.join(order_df, on=label_col, how="inner")
        .sort("sort_order")
        .drop("sort_order")
    )


plt.style.use("seaborn-v0_8-whitegrid")

# 1) Total demand over time.
daily_total_rows = daily_total_sales.to_dicts()
fig, ax = plt.subplots(figsize=(12, 4.5))
ax.plot(
    [row["date"] for row in daily_total_rows],
    [row["total_sales"] for row in daily_total_rows],
    color="#9aa0a6",
    linewidth=1,
    alpha=0.45,
    label="Daily total",
)
ax.plot(
    [row["date"] for row in daily_total_rows],
    [row["rolling_mean_10_service_days"] for row in daily_total_rows],
    color="#1f77b4",
    linewidth=2.5,
    label="10 service-day average",
)
ax.set_title("Total Daily Demand Over Time")
ax.set_xlabel("")
ax.set_ylabel("Dishes sold")
ax.legend(frameon=False)
fig.autofmt_xdate()
plt.show()

# 2) Weekday, season, and semester/academic-period effects.
weekday_effect = ordered_bar_data(
    daily_total_sales.group_by("weekday").agg(pl.mean("total_sales").alias("avg_sales")),
    "weekday",
    "avg_sales",
    weekday_order,
)
season_effect = ordered_bar_data(
    daily_total_sales.group_by("season").agg(pl.mean("total_sales").alias("avg_sales")),
    "season",
    "avg_sales",
    season_order,
)
academic_effect = ordered_bar_data(
    daily_total_sales.group_by("academic_bucket").agg(pl.mean("total_sales").alias("avg_sales")),
    "academic_bucket",
    "avg_sales",
    academic_bucket_order,
)
semester_effect = ordered_bar_data(
    daily_total_sales.group_by("semester").agg(pl.mean("total_sales").alias("avg_sales")),
    "semester",
    "avg_sales",
    semester_order,
)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
effect_specs = [
    (axes[0, 0], weekday_effect, "weekday", "Average Demand by Weekday"),
    (axes[0, 1], season_effect, "season", "Average Demand by Season"),
    (axes[1, 0], semester_effect, "semester", "Average Demand by Semester"),
    (axes[1, 1], academic_effect, "academic_bucket", "Average Demand by Academic Period"),
]

for ax, data, label_col, title in effect_specs:
    plot_rows = data.to_dicts()
    ax.bar(
        [row[label_col] for row in plot_rows],
        [row["avg_sales"] for row in plot_rows],
        color="#4c78a8",
    )
    ax.set_title(title)
    ax.set_xlabel("")
    ax.set_ylabel("Average dishes sold")
    ax.tick_params(axis="x", rotation=30)
    ax.bar_label(ax.containers[0], fmt="%.0f", padding=3)

fig.tight_layout()
plt.show()

# 3) Weekday effect split by lecture period and lecture-free period.
lecture_weekday_effect = (
    daily_total_sales
    .with_columns(
        period_type=pl.when(pl.col("is_lecture_day"))
        .then(pl.lit("Lecture period"))
        .otherwise(pl.lit("Lecture-free period"))
    )
    .group_by("weekday", "period_type")
    .agg(pl.mean("total_sales").alias("avg_sales"))
)

lecture_weekday_lookup = {
    (row["weekday"], row["period_type"]): row["avg_sales"]
    for row in lecture_weekday_effect.to_dicts()
}
lecture_values = [
    lecture_weekday_lookup.get((weekday, "Lecture period"), 0)
    for weekday in weekday_order
]
lecture_free_values = [
    lecture_weekday_lookup.get((weekday, "Lecture-free period"), 0)
    for weekday in weekday_order
]
x_positions = list(range(len(weekday_order)))
bar_width = 0.38

fig, ax = plt.subplots(figsize=(10, 5))
lecture_bars = ax.bar(
    [position - bar_width / 2 for position in x_positions],
    lecture_values,
    width=bar_width,
    label="Lecture period",
    color="#4c78a8",
)
lecture_free_bars = ax.bar(
    [position + bar_width / 2 for position in x_positions],
    lecture_free_values,
    width=bar_width,
    label="Lecture-free period",
    color="#f58518",
)
ax.set_title("Average Demand by Weekday: Lecture vs Lecture-Free Period")
ax.set_xlabel("")
ax.set_ylabel("Average dishes sold")
ax.set_xticks(x_positions)
ax.set_xticklabels(weekday_order, rotation=0)
ax.legend(frameon=False)
ax.bar_label(lecture_bars, fmt="%.0f", padding=3)
ax.bar_label(lecture_free_bars, fmt="%.0f", padding=3)
fig.tight_layout()
plt.show()

# 4) Category effects: total contribution over the year.
category_effect = (
    mock_sales_with_calendar
    .group_by("meal category")
    .agg(pl.sum("sales").alias("total_sales"))
    .with_columns(
        (pl.col("total_sales") / pl.sum("total_sales") * 100).alias("sales_share_pct")
    )
    .sort("total_sales")
)

category_effect_rows = category_effect.to_dicts()
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(
    [row["meal category"] for row in category_effect_rows],
    [row["sales_share_pct"] for row in category_effect_rows],
    color="#59a14f",
)
ax.set_title("Share of Annual Demand by Category")
ax.set_xlabel("Share of dishes sold (%)")
ax.set_ylabel("")
ax.bar_label(ax.containers[0], fmt="%.1f%%", padding=3)
fig.tight_layout()
plt.show()

daily_total_sales.head()
```
